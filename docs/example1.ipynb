# 1. Individual signals

```python
import matplotlib.pyplot as plt
from pypqd import synthetic
```

```python
signal = synthetic.sag()
time = signal[0]["Time"]
data = signal[0]["Values"]
plt.figure(figsize=(8, 1.5))
plt.plot(time, data)
plt.grid()
title_line1 = signal[0]["Label"]
title_line2 = "\n (Sag magnitude: {:.2f})".format(signal[0]["Sag magnitude"])
plt.title(title_line1 + title_line2, fontsize=10)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude [pu]")
plt.show()
```

```python
signal = synthetic.interruption(interruption_magnitude = (0.95,), interruption_interval=("fixed", (0.05, 0.15)))
time = signal[0]["Time"]
data = signal[0]["Values"]
plt.figure(figsize=(8, 1.5))
plt.plot(time, data)
plt.grid()
title_line1 = signal[0]["Label"]
title_line2 = "\n (Interruption magnitude: {:.2f})".format(signal[0]["Interruption magnitude"])
plt.title(title_line1 + title_line2, fontsize=10)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude [pu]")
plt.show()
```
